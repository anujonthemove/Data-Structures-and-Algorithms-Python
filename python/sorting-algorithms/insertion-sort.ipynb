```python
# My original thought to solving this
def insertion_sort(arr, debug=False):
    for i in range(1, len(arr)):
        elem = arr[i]
        idx = i
        
        while(elem < arr[idx-1]):
            arr[idx] = arr[idx-1]
            if idx > 0: idx -=1
            else: break
        arr[idx] = elem
    return arr
```
